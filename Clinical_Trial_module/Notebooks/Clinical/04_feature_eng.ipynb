```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/clinical_processed/cleaned_data.csv")

# simplify phase
df['phase'] = df['phases'].astype(str).str.extract('(\d)').astype(float)

# log transform
df['log_enrollment'] = df['enrollment'].apply(lambda x: np.log1p(x))

# final columns
df = df[[
    'phase',
    'log_enrollment',
    'duration_months',
    'study_type',
    'funder_type',
    'target'
]]

df.to_csv("../../data/clinical_processed/feature_data.csv", index=False)

print("Feature data ready:", df.shape)
```
