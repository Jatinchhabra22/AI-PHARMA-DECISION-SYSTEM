```python
import pandas as pd
import numpy as np

df = pd.read_csv("../../data/clinical_raw/ctg-studies.csv")

# standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# keep useful columns
df = df[[
    'study_status',
    'phases',
    'enrollment',
    'study_type',
    'funder_type',
    'start_date',
    'completion_date'
]]

# target creation
df['target'] = df['study_status'].apply(
    lambda x: 1 if str(x).lower() == 'completed' else 0
)

# convert dates
df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
df['completion_date'] = pd.to_datetime(df['completion_date'], errors='coerce')

# duration
df['duration_months'] = (df['completion_date'] - df['start_date']).dt.days / 30

# clean
df = df.dropna()

df.to_csv("../../data/clinical_processed/cleaned_data.csv", index=False)

print("Cleaned data saved:", df.shape)
```
