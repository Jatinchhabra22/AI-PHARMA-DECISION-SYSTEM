```python
import pandas as pd

df = pd.read_csv("../data/Final_data/drug_final_enriched_dataset.csv")
```

```python
def get_drug_insights(drug_name):
    
    drug = df[df['drugName'].str.lower() == drug_name.lower()]
    
    if drug.empty:
        return "Drug not found"
    
    drug = drug.iloc[0]
    
    result = {
        "Drug Name": drug['drugName'],
        
        "Sentiment Score": round(drug['avg_sentiment'], 3),
        
        "Predicted Demand": round(drug['demand'], 2),
        
        "Effectiveness Score": round(drug['effectiveness_score'], 2),
        
        "Category": drug['drug_category'],
        
        "Top Side Effects": drug['top_side_effects'],
        
        "Insights": drug['insights'],
        
        "Recommended Actions": drug['actions']
    }
    
    return result
```

```python
get_drug_insights("Abilify")
```

```python
def search_similar_drugs(category):
    
    return df[df['drug_category'] == category][
        ['drugName', 'performance_score']
    ].sort_values(by='performance_score', ascending=False).head(5)
```

```python
search_similar_drugs("High Performer")
```
